```python
import pandas as pd
```

```python
products = pd.read_csv("data/item_metadata_filtered.csv")
```

```python
products
```

```python
import matplotlib.pyplot as plt

# Filter to only rows where images are available
df_images = products[products['has_images'] == True]

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df_images['image_count'], bins=range(1, df_images['image_count'].max() + 2), edgecolor='black')
plt.title("Distribution of Image Counts per Product")
plt.xlabel("Number of Images")
plt.ylabel("Number of Products")
plt.xticks(range(1, df_images['image_count'].max() + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
